```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, random_split
import numpy as np
import pandas as pd

# 自定义数据集类（保持与原始代码一致）
class SensorDataset(Dataset):
    def __init__(self, data_path):
        data = pd.read_csv(data_path).values
        self.X = data[:, :84].astype(np.float32)
        self.y = data[:, -42:].astype(np.float32)
        
    def __len__(self):
        return len(self.X)
    
    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]

# 基于1D卷积的变分自编码器
class Conv1DVAE(nn.Module):
    def __init__(self, latent_dim=16):
        super().__init__()
        self.latent_dim = latent_dim
        
        # 编码器
        self.encoder = nn.Sequential(
            nn.Conv1d(1, 32, kernel_size=3, padding=1),
            nn.BatchNorm1d(32),
            nn.ReLU(),
            nn.MaxPool1d(2),
            nn.Conv1d(32, 16, kernel_size=3, padding=1),
            nn.BatchNorm1d(16),
            nn.ReLU(),
            nn.MaxPool1d(2)
        )
        
        # 潜在空间均值和对数方差
        self.fc_mu = nn.Linear(16*21, latent_dim)  # 16 channels * 21 time steps
        self.fc_logvar = nn.Linear(16*21, latent_dim)
        
        # 解码器
        self.decoder = nn.Sequential(
            nn.Linear(latent_dim, 16*21),
            nn.Unflatten(1, (16, 21)),
            nn.ConvTranspose1d(16, 32, kernel_size=2, stride=2),
            nn.BatchNorm1d(32),
            nn.ReLU(),
            nn.ConvTranspose1d(32, 1, kernel_size=2, stride=2),
            nn.BatchNorm1d(1),
            nn.ReLU(),
            nn.Linear(84, 42)
        )
    
    def reparameterize(self, mu, logvar):
        std = torch.exp(0.5*logvar)
        eps = torch.randn_like(std)
        return mu + eps*std
    
    def forward(self, x):
        # 编码
        x = x.unsqueeze(1)
        encoded = self.encoder(x)
        encoded = encoded.view(encoded.size(0), -1)
        
        mu = self.fc_mu(encoded)
        logvar = self.fc_logvar(encoded)
        z = self.reparameterize(mu, logvar)
        
        # 解码
        decoded = self.decoder(z)
        return decoded.squeeze(1), mu, logvar

# 修改后的训练配置
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
batch_size = 64
latent_dim = 16
learning_rate = 1e-3
kl_weight = 0.001  # KL散度权重系数
num_epochs = 100

# 数据加载（保持与原始代码一致）
dataset = SensorDataset("log.csv")
train_size = int(0.8 * len(dataset))
train_dataset, val_dataset = random_split(dataset, [train_size, len(dataset)-train_size])

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size)

# 初始化VAE模型
model = Conv1DVAE(latent_dim=latent_dim).to(device)
optimizer = optim.Adam(model.parameters(), lr=learning_rate)

# VAE损失函数
def vae_loss(recon_x, x, mu, logvar):
    recon_loss = nn.MSELoss(reduction='sum')(recon_x, x)
    kl_div = -0.5 * torch.sum(1 + logvar - mu.pow(2) - logvar.exp())
    return recon_loss + kl_weight * kl_div

# 修改后的训练循环
best_val_loss = float('inf')
train_losses = []
for epoch in range(num_epochs):
    # 训练阶段
    model.train()
    train_loss = 0
    for inputs, targets in train_loader:
        inputs, targets = inputs.to(device), targets.to(device)
        
        recon_batch, mu, logvar = model(inputs)
        loss = vae_loss(recon_batch, targets, mu, logvar)
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        train_loss += loss.item()
    
    # 验证阶段
    model.eval()
    val_loss = 0
    with torch.no_grad():
        for inputs, targets in val_loader:
            inputs, targets = inputs.to(device), targets.to(device)
            recon, mu, logvar = model(inputs)
            val_loss += vae_loss(recon, targets, mu, logvar).item()
    
    train_loss /= len(train_loader.dataset)
    val_loss /= len(val_loader.dataset)
    train_losses.append(train_loss)
    
    # 保存最佳模型
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        torch.save(model.state_dict(), 'best_Conv1DVAE_model.pth')
    
    print(f'Epoch [{epoch+1}/{num_epochs}] | Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f}')
```

```python
# ========================== 5. 可视化学习曲线 ========================== #
plt.plot(train_losses, label='Train Loss')
plt.legend()
plt.title('Training loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()
```
